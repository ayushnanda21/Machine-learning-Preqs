```python
titanic_data = pd.read_csv('https://raw.githubusercontent.com/zekelabs/data-science-complete-tutorial/master/Data/titanic-train.csv.txt')
```

```python
titanic_data.info()
```

```python
#To drop a column
titanic_data.drop(['Cabin'], axis=1, inplace=True)
```

```python
titanic_data.info()
```

```python
# if we want to drop all columns in which any data is missing use (dropna)
titanic_data.dropna().info()
```

```python
#if columns specified in subset are empty they will be dropped else not
titanic_data.dropna(subset=['Embarked', 'Age']).info()
```

```python
# to fill missing data
titanic_data.fillna({'Age':0,'Embarked':'Unknown'}).info()
```

```python
titanic_data.Age.fillna(method='ffill')[:5]
```

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3','K4','K5'],
                         'A': ['A0', 'A1', 'A2', 'A3','A4','A5'],
                         'B': ['B0', 'B1', 'B2', 'B3','B4','B5']})
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3','K6','K7'],
                          'C': ['C0', 'C1', 'C2', 'C3','C6','C7'],
                         'D': ['D0', 'D1', 'D2', 'D3','D6','D7']})
```

```python
left.merge(right, on='key')
```

```python
left.merge(right, on='key', how='left')
```

```python
#Combining data based on inner joints
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                     index=['K0', 'K1', 'K2'])

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                      'D': ['D0', 'D2', 'D3']},
                     index=['K0', 'K2', 'K3'])
```

```python
left.join(right)
```
