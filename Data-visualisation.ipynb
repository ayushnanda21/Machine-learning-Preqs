```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Line PLot
```

```python
x = np.arange(10)
y1 = x**2
y2 = 2*x+3
print(x)
print(y1)
print(y2)
```

```python
themes = plt.style.available
print(themes)
```

```python
plt.style.use("seaborn")
```

```python
plt.plot(x,y1,color="red",label="Apple",marker='o')
plt.plot(x,y2,label="Kiwi",linestyle="dashed",marker='*')
plt.legend()
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of Fruits over Time ")
plt.show()
```

```python
prices = np.array([1,2,3,4])**2
plt.plot(prices)
plt.show()
```

```python
##SCATTER PLOTS
```

```python
#Adjusting size of plot : 
plt.figure(figsize=(5,5))
plt.scatter(x,y1,color="red",label="Apple",marker='o')
plt.scatter(x,y2,label="Kiwi",linestyle="dashed",marker='*')
plt.legend()
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Scatter PLot Prices of Fruits over Time ")
plt.show()
```

```python
# BAR GRAPHS
```

```python
plt.bar([1,2,3,4],[1,5,11,17])
plt.show()
```

```python
x_coordinates = np.array([1,2,3,4])**2
plt.bar(x_coordinates,[10,20,15,18],width=0.5,label="2020",tick_label=["Gold", "Iron","Silver","Nickel"],color="blue")#year 2020
plt.bar(x_coordinates+0.5,[10,11,18,16],width=0.5,label="2021",color="black")#year 2021
plt.ylim(0,40)
plt.xlabel("Metals")
plt.ylabel("Price")
plt.title("Metals price comparison")
plt.legend()
plt.show()
```

```python
#PIE CHARTS
```

```python
subjects= "Maths","Chemistry","Physics","English"
weightage = [20,10,15,5]
plt.pie(weightage, explode=None, labels=subjects,  autopct='%1.1f%%', shadow=True)
plt.show()
```

```python
#HISTOGRAM
```

```python
Xsn = np.random.randn(100)  #normal distribution

sigma = 60
u=5
X= Xsn*sigma + u
X2= np.round(Xsn*5+40)
plt.style.use("seaborn")
plt.title("Histogram")
plt.xlabel("Marks")
plt.ylabel("Probability/Frequeny")
plt.hist(X)
plt.hist(X2)
plt.show()

```
