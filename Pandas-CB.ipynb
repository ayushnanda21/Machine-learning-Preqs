```python
 # PANDAS MOVIE DATASET TASK
#Task : Make scatter plot between length of movie title and freq of that length
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("movie_metadata.csv")
df
```

```python
df.head()
```

```python
df.columns
```

```python
titles = list(df.get('movie_title'))

```

```python
freq_title ={}

for t in titles:
    length= len(t)
    print(t)
    print(length)
    
    try:
        freq_title[length] += 1
    except:
        freq_title[length] =1
        
print(freq_title)
        
```

```python
X= np.array(list(freq_title.keys()))
Y= np.array(list(freq_title.values()))
```

```python
plt.scatter(X,Y)
plt.xlabel("Movie title length")
plt.ylabel("No. of movies")
plt.show()
```
